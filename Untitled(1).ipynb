## 1. Ojetivo Trabajo
El objetivo de este trabajo es analizar estadisticamente el tiempo de corrida de un algo
ritmo matemeatico y determinar como cambia en funci´on de la carga de tarea pedida.

## 2. Introduccion
La forma mas sencilla de medir el tiempo de ejecucion de un algoritmo es utilizando
funciones provistas por el sistema operativo o el lenguaje de programacion que permiten
obtener la hora actual del sistema antes y despues de la ejecucion del algoritmo. La diferencia
entre ambos tiempos nos da una estimacion del tiempo que tardo en ejecutarse el algoritmo.

## 3.Estadistica Descriptiva


Antes que nada queremos entender los datos que recolectamos y hacernos preguntas, como por ejemplo si el tiempo de computo es una variable aleatoria o es determinista. A continuación se pueden observar gráficos muestran lo mencionado.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_json("bench_fib.json")
#Queremos evaluar si el tiempo de ejecucion es una variable aleatoria. Es decir que su valor no es determinista
#Evaluemos primero para cierto n elejido elegido.

df1 = df[(df["n"] == 10) & (df["batch_size"] == 5)].sample(n=500, random_state=42)
df2 = df[(df["n"] == 10) & (df["batch_size"] == 50)].sample(n=500, random_state=42)
df3 = df[(df["n"] == 30) & (df["batch_size"] == 5)].sample(n=500, random_state=42)
df4 = df[(df["n"] == 30) & (df["batch_size"] == 50)].sample(n=500, random_state=42)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].hist(df1["time_ms"], 30, alpha=0.8, label='batch size = 5', color='red')
axs[0].hist(df2["time_ms"], 30, alpha=0.8, label='batch size = 50', color='green')
axs[0].set_xlabel("Tiempo de computo (ms)")
axs[0].set_ylabel("Apariciones")
axs[0].set_title("n = 10")
axs[0].legend()

axs[1].hist(df3["time_ms"], 30, alpha=0.8, label='batch size = 5', color='red')
axs[1].hist(df4["time_ms"], 30, alpha=0.8, label='batch size = 50', color='green')
axs[1].set_xlabel("Tiempo de computo (ms)")
axs[1].set_ylabel("Apariciones")
axs[1].set_title("n = 30")
axs[1].legend()

fig.tight_layout()
plt.show()
```

El primer gráfico, con n = 10, muestra una distribución del tiempo concentrada en valores más bajos, tanto para un batch size de 5 como de 50. Los casos en los que el tiempo supera los 0.0075 ms son muy pocos. Por otro lado, para un n mayor, como en el segundo gráfico, se observa una distribución que tiende a ser más normal, con valores centrados alrededor de 112.5 ms.

Los sistemas computacionales, si bien son deterministas, poseen una gran cantidad de variables, subprocesos y estados posibles; por ello, ejecutar una función en distintos momentos puede arrojar tiempos de ejecución diferentes. Dependiendo de los procesos que esté realizando el procesador, el tiempo de cómputo de la función de Fibonacci puede verse afectado.

A medida que tarda más en calcular la función, aumenta la probabilidad de que haya mayores diferencias entre los subprocesos, lo que genera una distribución menos concentrada y con forma de campana. Como conclusión, dado que el tiempo de ejecución no es determinista, podemos considerarlo una variable aleatoria.

```python
df1 = df[(df["n"] == 4) & (df["batch_size"] == 1)].sample(n=500, random_state=42)
df2 = df[(df["n"] == 4) & (df["batch_size"] == 5)].sample(n=500, random_state=42)
df3 = df[(df["n"] == 4) & (df["batch_size"] == 10)].sample(n=500, random_state=42)
df4 = df[(df["n"] == 4) & (df["batch_size"] == 50)].sample(n=500, random_state=42)

df5 = df[(df["n"] == 20) & (df["batch_size"] == 1)].sample(n=500, random_state=42)
df6 = df[(df["n"] == 20) & (df["batch_size"] == 5)].sample(n=500, random_state=42)
df7 = df[(df["n"] == 20) & (df["batch_size"] == 10)].sample(n=500, random_state=42)
df8 = df[(df["n"] == 20) & (df["batch_size"] == 50)].sample(n=500, random_state=42)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].hist(df1["time_ms"], 30, alpha=0.8, label='batch size = 1', color='red')
axs[0].hist(df2["time_ms"], 30, alpha=0.8, label='batch size = 5', color='green')
axs[0].hist(df3["time_ms"], 30, alpha=0.8, label='batch size = 10', color='blue')
axs[0].hist(df4["time_ms"], 30, alpha=0.8, label='batch size = 50', color='yellow')
axs[0].set_xlabel("Tiempo de computo (ms)")
axs[0].set_ylabel("Apariciones")
axs[0].set_title("n = 10")
axs[0].legend()

axs[1].hist(df5["time_ms"], 30, alpha=0.8, label='batch size = 1', color='red')
axs[1].hist(df6["time_ms"], 30, alpha=0.8, label='batch size = 5', color='green')
axs[1].hist(df7["time_ms"], 30, alpha=0.8, label='batch size = 10', color='blue')
axs[1].hist(df8["time_ms"], 30, alpha=0.8, label='batch size = 50', color='yellow')
axs[1].set_xlabel("Tiempo de computo (ms)")
axs[1].set_ylabel("Apariciones")
axs[1].set_title("n = 30")
axs[1].legend()

fig.tight_layout()
plt.show()
```

## Mediciones y estadisticos Interesantes

Seguiendo con la estadistica descriptiva nos interesa comparar estadisticos entre los distintos batchs. Para ello a continuacion se puede ver una tabla con dichas mediciones.

```python
import numpy as np

batch_sizes = [1, 5, 10, 50]
ns = [2, 4, 10, 20, 30]
resultados = []
for batch_size in batch_sizes:
    for n in ns:
        df1 = df[(df["n"] == n) & (df["batch_size"] == batch_size)]
        datos = np.array(df1["time_ms"])
        media = datos.sum() / len(datos)
        desviacion_estandar = np.std(np.array(datos))
        mediana = np.median(datos)
        intercuartil = np.percentile(datos, 75) - np.percentile(datos, 25)

        resultados.append({
        "n": n,
        "batch size": batch_size,
        "media": media,
        "desviacion estandar": desviacion_estandar,
        "mediana": mediana,
        "intercuartil": intercuartil
    })

# Convertir a tabla
tabla = pd.DataFrame(resultados)
tabla = tabla.sort_values(by=["n", "batch size"])
print(tabla.to_string(index=False))
```

Nuevamente, y tal como se observa en los gráficos, la desviación estándar tiende a ser mayor para los valores de n superiores a 30. 

## Mediciones y estadisticos Interesantes
Nos interesa poder testear si efectivamente las distribuciones que obtuvimos provienen de una distribucion normal como intuiamos a partir de los graficos. Realizaremos tests de normalidad para poder llegar a una conclusion basada en los datos.
Planteamos las siguientes hipotesis:

H0 = la distribucion no proviene de una muestra normal
H1 = la distribucion proviene de una muestra normal

Usaremos el test de normalidad de Shapiro-Wilks eligiendo un nivel de confianza estandar del 0.05.

```python
from scipy.stats import shapiro, kstest, norm


n = 30
batch_size = 50

data = df[(df["n"] == n) & (df["batch_size"] == batch_size)]["time_ms"].values

# Shapiro 
W, p_shapiro = shapiro(data)
print("Shapiro p:", p_shapiro)

# Kolmogorov con normal ajustada
mu, sigma = np.mean(data), np.std(data, ddof=1)
stat, p_ks = kstest(data, 'norm', args=(mu, sigma))
print("KS p:", p_ks)
```
