## 1.Ojetivo Trabajo
El objetivo de este trabajo es analizar estadisticamente el tiempo de corrida de un algo
ritmo matemeatico y determinar como cambia en funci´on de la carga de tarea pedida.

## 2. Introduccion
La forma mas sencilla de medir el tiempo de ejecucion de un algoritmo es utilizando
funciones provistas por el sistema operativo o el lenguaje de programacion que permiten
obtener la hora actual del sistema antes y despues de la ejecucion del algoritmo. La diferencia
entre ambos tiempos nos da una estimacion del tiempo que tardo en ejecutarse el algoritmo.

## 3.Estadistica Descriptiva


```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_json("bench_fib.json")
#Queremos evaluar si el tiempo de ejecucion es una variable aleatoria. Es decir que su valor no es determinista
#Evaluemos primero para cierto n elejido elegido.

df1 = df[(df["n"] == 10) & (df["batch_size"] == 5)].sample(n=500, random_state=42)
df2 = df[(df["n"] == 10) & (df["batch_size"] == 50)].sample(n=500, random_state=42)
df3 = df[(df["n"] == 30) & (df["batch_size"] == 5)].sample(n=500, random_state=42)
df4 = df[(df["n"] == 30) & (df["batch_size"] == 50)].sample(n=500, random_state=42)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].hist(df1["time_ms"], 30, alpha=0.8, label='Dataset 1', color='red')
axs[0].hist(df2["time_ms"], 30, alpha=0.8, label='Dataset 1', color='green')
axs[0].set_xlabel("Tiempo de computo (ms)")
axs[0].set_ylabel("Apariciones")
axs[0].set_title("n = 10")

axs[1].hist(df3["time_ms"], 30, alpha=0.8, label='Dataset 1', color='red')
axs[1].hist(df4["time_ms"], 30, alpha=0.8, label='Dataset 1', color='green')
axs[1].set_xlabel("Tiempo de computo (ms)")
axs[1].set_ylabel("Apariciones")
axs[1].set_title("n = 30")

fig.tight_layout()
plt.show()
```
